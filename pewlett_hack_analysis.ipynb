````python
#Set the dependancies
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
from sqlalchemy.types import Integer,Text, String, Float
import psycopg2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
````

````python
engine = create_engine('postgresql://postgres:PASSWORD@localhost:5432/pewlett_hack')
connection = engine.connect()
````

````python
#Query the salaries in the salaries table
query1 = "select * from salaries"
salary_df = pd.read_sql(query1, connection)
salary_df.head(20)
````

Group the salaries into bands

````python
#Create the bins
salary_bins = [39999.99, 49999.99, 59999.99, 69999.99, 79999.99, 89999.99, 99999.99, 109999.99, 119999.99, 130000 ]
````

````python
#Create the salary labels for the salary bins
salary_labels = ['$40,000 - $49,999',
                '$50,000 - $59,999',
                '$60,000 - $69,999',
                '$70,000 - $79,999',
                '$80,000 - $89,999',
                '$90,000 - $99,999',
                '$100,000 - $109,999',
                '$110,000 - $119,999',
                '$120,000 - $130,000']
````

````python
#Slice the data and place it into bins, put into the salary dataframe
salary_df["salary_band"] = pd.cut(salary_df['salary'], salary_bins, labels=salary_labels)
````

````python
#Do a groupby to condence the data
grouped_salary = salary_df.groupby(['salary_band']).count()
grouped_salary
````

````python
grouped_salary.reset_index()
````

````python
#Create a histogram to visualize the most common salary ranges for employees.
plt.figure(figsize=(10,6))
grouped_salary.plot.bar(align="center", color="orange")
plt.title('Number of employees per salary band\n Pewlett Hackard\n')
plt.ylabel('Number of Employees')
plt.xlabel('Salary Bands ($)')
plt.grid()
plt.legend().remove()
plt.show()
````

````python
#Query the salaries in the titles table
query2 = "select * from titles"
titles_df = pd.read_sql(query2, connection)
titles_df.head()
````

````python
#Query the employees table to merge
query3 = "select * from employees"
employees_df = pd.read_sql(query3, connection)
employees_df.head()
````

````python
#Merge the employee table with the salary table
employee_salaries_merge = pd.merge(employees_df, salary_df, on='emp_no')
````

````python
#Change the header in the employee_salaries_merge dataframe
employee_salaries_merge2 = employee_salaries_merge.rename(columns={'emp_title_id': 'title_id'})
````

````python
#Merge employee_salaries_merge with the titles table
full_employee_details = pd.merge(employee_salaries_merge2, titles_df, on='title_id')
````

````python
full_employee_details.head(10)
````

````python
#Group and calculate the average salary by title
salary_title = full_employee_details.groupby("title")["salary"].mean()
````

````python
salary_title
````

````python
#Create a bar chart of average salary by title.
plt.figure(figsize=(10,6))
salary_title.plot.bar(align="center", color="orange")
plt.title('Avaerage Salary by Job Title\n')
plt.xlabel('Title')
plt.ylabel('Average Salary ($)')
plt.grid()
plt.legend().remove()
plt.show()
````
